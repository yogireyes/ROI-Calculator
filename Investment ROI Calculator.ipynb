```python
class Property:
    def __init__(self, rental_income, laundry_income, storage_income, miscellaneous_income,
                 tax, insurance, utilities, hoa, outdoor_maintenance, vacancy, repairs,
                 long_term_expenditures, property_manager, mortgage):
        self.rental_income = rental_income
        self.laundry_income = laundry_income
        self.storage_income = storage_income
        self.miscellaneous_income = miscellaneous_income
        self.tax = tax
        self.insurance = insurance
        self.utilities = utilities
        self.hoa = hoa
        self.outdoor_maintenance = outdoor_maintenance
        self.vacancy = vacancy
        self.repairs = repairs
        self.long_term_expenditures = long_term_expenditures
        self.property_manager = property_manager
        self.mortgage = mortgage

    def total_monthly_income(self):
        return sum([self.rental_income, self.laundry_income, self.storage_income, self.miscellaneous_income])

    def total_monthly_expenses(self):
        return sum([self.tax, self.insurance, self.utilities, self.hoa, self.outdoor_maintenance,
                    self.vacancy, self.repairs, self.long_term_expenditures, self.property_manager, self.mortgage])

class Investment:
    def __init__(self, property, down_payment, closing_cost, rehab_budget, miscellaneous_investment):
        self.property = property
        self.down_payment = down_payment
        self.closing_cost = closing_cost
        self.rehab_budget = rehab_budget
        self.miscellaneous_investment = miscellaneous_investment

    def total_investment(self):
        return self.down_payment + self.closing_cost + self.rehab_budget + self.miscellaneous_investment

    def calculate_cash_on_cash_roi(self):
        annual_cash_flow = (self.property.total_monthly_income() - self.property.total_monthly_expenses()) * 12
        return (annual_cash_flow / self.total_investment()) * 100

# Gathering user input
print("Welcome to the Rental Property ROI Calculator!")

try:
    rental_income = float(input("Enter monthly rental income ($): "))
    laundry_income = float(input("Enter monthly laundry income ($): "))
    storage_income = float(input("Enter monthly storage income ($): "))
    miscellaneous_income = float(input("Enter monthly miscellaneous income ($): "))

    tax = float(input("Enter monthly tax expense ($): "))
    insurance = float(input("Enter monthly insurance expense ($): "))
    utilities = float(input("Enter monthly utilities expense ($): "))
    hoa = float(input("Enter monthly HOA fee ($): "))
    outdoor_maintenance = float(input("Enter monthly outdoor maintenance expense ($): "))
    vacancy = float(input("Enter monthly vacancy expense ($): "))
    repairs = float(input("Enter monthly repairs expense ($): "))
    long_term_expenditures = float(input("Enter monthly long term expenditures expense ($): "))
    property_manager = float(input("Enter monthly property manager fee ($): "))
    mortgage = float(input("Enter monthly mortgage expense ($): "))

    down_payment = float(input("Enter down payment amount ($): "))
    closing_cost = float(input("Enter closing cost amount ($): "))
    rehab_budget = float(input("Enter rehab budget amount ($): "))
    miscellaneous_investment = float(input("Enter miscellaneous investment amount ($): "))

    # Calculating ROI
    property = Property(rental_income, laundry_income, storage_income, miscellaneous_income,
                        tax, insurance, utilities, hoa, outdoor_maintenance, vacancy, repairs,
                        long_term_expenditures, property_manager, mortgage)
    investment = Investment(property, down_payment, closing_cost, rehab_budget, miscellaneous_investment)
    cash_on_cash_roi = investment.calculate_cash_on_cash_roi()

    print(f"Cash on Cash ROI: {cash_on_cash_roi:.2f}%")
except ValueError:
    print("Invalid input. Please enter numerical values.")
```
